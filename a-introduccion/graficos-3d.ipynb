## Trazado de Espiral

```python
"""Trazado de espiral.

trazado_espiral_3d.py

"""

fig = plt.figure()
ax = fig.gca(projection='3d')

t = np.linspace(0,20, 100)
x = t * np.sin(t);
y = t * np.sin(t);
z = t;

ax.plot(x, y, z, label='Espiral')
ax.legend()

plt.show()
```
