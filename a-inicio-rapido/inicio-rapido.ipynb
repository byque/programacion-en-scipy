Cuando se inicia Jupyter Notebook desde la consola, el uso de `--pylab inline` es importante para que los gráficos se muestren dentro del notebook y no en una nueva ventana. Alternativamente se puede usar el siguiente código dentro de un notebook para lograr el mismo resultado.

```python
u"""Inicio rápido.

inico_rapido.py

Computar el máximo de una función de Bessel y graficarlo.

"""

import numpy as np
from scipy import special, optimize
import matplotlib.pyplot as plt

# Compute maximum
f = lambda x: -special.jv(3, x)
sol = optimize.minimize(f, 1.0)

# Plot
x = np.linspace(0, 10, 5000)
plt.plot(x, special.jv(3, x), '-', sol.x, -sol.fun, 'o')
```
